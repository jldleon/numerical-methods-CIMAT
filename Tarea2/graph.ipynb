```python
from math import exp as exp

def newton(p0, TOL, N):
    
    for i in range(1, N+1):
        p = p0 - f(p0)/fder(p0)
        
        if (abs(p-p0) < TOL ):
            return p
        
        p0 = p
        
    return "FALLO EL METODO"

def f(x):
    return 1000000*exp(x)+(435000/x)*(exp(x)-1)-1564000
    

def fder(x):
    return 1000000*exp(x)+435000*(x*exp(x)-exp(x)+1)/(x*x)

x_punto = newton(p0 = 0.1, TOL = 0.0001, N=20)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 100)

y = []

for i in x:
    y.append(f(i))


y_punto = 0.0

plt.plot(x, y)

plt.scatter(x_punto, y_punto, color='red')


# Muestra el gráfico
plt.grid(True)
plt.show()
```
